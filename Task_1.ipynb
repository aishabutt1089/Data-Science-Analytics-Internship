# Task 1: Exploring and Visualizing Iris Dataset
**Objective:** To understand how to read, summarize, and visualize a dataset using pandas, matplotlib, and seaborn.

```python
# Essential Libraries Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset using seaborn
df = sns.load_dataset('iris')
```

```python
# 2. Display dataset structure
print("--- Dataset Shape ---")
print(df.shape)  # Rows and columns count

print("\n--- Dataset Columns ---")
print(df.columns)  # Column names

print("\n--- First 5 Rows ---")
print(df.head())  # Preview data

# 3. Data Summarization
print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
# --- Visualizations ---
# Set theme
sns.set_theme(style="whitegrid")

# Plot 1: Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Plot 2: Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='petal_length', kde=True, hue='species', bins=20, multiple="stack")
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.show()

# Plot 3: Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal_width', hue='species', palette='Set2', legend=False)
plt.title('Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.show()
```

### Conclusion:
- **Iris-setosa** occupies a distinct cluster in sepal measurements.
- **Petal length and width** are strong predictors for differentiating the three species.
